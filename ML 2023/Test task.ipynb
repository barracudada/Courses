6. Имеется две строки $string_1$ и $string_2$. Выыедите конкатенацию этих строк, написанную наоборот. Пример: $string_1 = house$, $string_2 = cat$, результат: tacesuoh. 

```python
string_1 = 'house'
string_2 = 'cat'

def reversed1(string1, string2):
    res = ''
    string = string1 + string2
    for i in range(len(string)-1,-1,-1):
        res+=string[i]
    return(res)

def reversed2(string1, string2):
    string = string1 + string2
    def subf(stringg):
        if len(stringg) == 1:
            return(stringg)
        else:
            return(stringg[-1]+ subf(stringg[:-1]))
    string = subf(string)
    return(string)

def reversed3(string1, string2):
    res = ''.join(reversed(string1 + string2))
    return(res)

def reversed4(string1, string2):
    return((string1+string2)[::-1])

reversed1(string_1, string_2), reversed2(string_1, string_2), reversed3(string_1, string_2), reversed4(string_1, string_2)
```
