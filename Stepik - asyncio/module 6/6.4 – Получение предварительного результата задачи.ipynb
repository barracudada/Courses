## wait_for

```python
import asyncio

async def slow():
    await asyncio.sleep(3)
    return 42

async def main():
    task = asyncio.create_task(slow())
    try:
        await asyncio.wait_for(task, timeout=1)
    except asyncio.TimeoutError:
        print('timeout')
    await task
    print(task.result())

asyncio.run(main())
```
