## add_done_callback / remove_done_callback

```python
import asyncio

def on_done(task):
    print('Callback: результат =', task.result())

async def work():
    await asyncio.sleep(1)
    return 123

async def main():
    task = asyncio.create_task(work())
    task.add_done_callback(on_done)
    await task

asyncio.run(main())
```
