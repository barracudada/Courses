## Конкурентное выполнение задач

```python
import asyncio

async def a():
    await asyncio.sleep(2)
    print('A done')

async def b():
    await asyncio.sleep(3)
    print('B done')

async def main():
    await asyncio.gather(asyncio.create_task(a()), asyncio.create_task(b()))

asyncio.run(main())
```
