## asyncio.shield — защита от отмены

```python
import asyncio

async def important():
    try:
        await asyncio.shield(asyncio.sleep(3))
        print('Критическая операция завершена')
    except asyncio.CancelledError:
        print('Отмена перехвачена')

async def main():
    task = asyncio.create_task(important())
    await asyncio.sleep(1)
    task.cancel()
    await task

asyncio.run(main())
```
