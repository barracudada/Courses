# 9.2.1 Проект - решение

```

import asyncio

robot_names = ['Электра', 'Механикс', 'Оптимус', 'Симулакр', 'Футуриус']
lock = asyncio.Lock()
counter = 0

async def robot(name, rid):
    global counter
    async with lock:
        print(f"Робот {name}({rid}) передвигается к месту A")
        await asyncio.sleep(1)
        counter += 1
        print(f"Робот {name}({rid}) достиг места A. Место A посещено {counter} раз")

async def main():
    tasks = [robot(name, i) for i, name in enumerate(robot_names)]
    await asyncio.gather(*tasks)

asyncio.run(main())
```
