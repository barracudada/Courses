## Защита ресурса с asyncio.Lock

```python
import asyncio

lock = asyncio.Lock()

async def worker(i):
    async with lock:
        print(f'Worker {i} вошел')
        await asyncio.sleep(1)
        print(f'Worker {i} вышел')

async def main():
    await asyncio.gather(*(worker(i) for i in range(3)))

asyncio.run(main())
```
