## asyncio.Condition

```python
import asyncio

condition = asyncio.Condition()

async def consumer():
    async with condition:
        print('Consumer ждет')
        await condition.wait()
        print('Consumer продолжает')

async def producer():
    await asyncio.sleep(1)
    async with condition:
        condition.notify_all()

async def main():
    await asyncio.gather(consumer(), producer())

asyncio.run(main())
```
