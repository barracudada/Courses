## asyncio.Barrier

```python
import asyncio

async def worker(i, barrier):
    print(f'Worker {i} ждет')
    await barrier.wait()
    print(f'Worker {i} продолжает')

async def main():
    barrier = asyncio.Barrier(3)
    await asyncio.gather(*(worker(i, barrier) for i in range(3)))

asyncio.run(main())
```
