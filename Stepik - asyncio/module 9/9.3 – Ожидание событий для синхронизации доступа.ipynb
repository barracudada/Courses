## asyncio.Event

```python
import asyncio

event = asyncio.Event()

async def waiter():
    print('Ждем событие')
    await event.wait()
    print('Событие произошло')

async def setter():
    await asyncio.sleep(1)
    event.set()

async def main():
    await asyncio.gather(waiter(), setter())

asyncio.run(main())
```
