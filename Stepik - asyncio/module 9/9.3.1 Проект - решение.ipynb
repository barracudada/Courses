# 9.3.1 Проект - решение

```

import asyncio

ip = ["192.168.0.3", "192.168.0.1", "192.168.0.2", "192.168.0.4", "192.168.0.5"]
alarm = asyncio.Event()

async def sensor(i, ip_addr):
    print(f"Датчик {i} IP-адрес {ip_addr} настроен и ожидает срабатывания")
    await alarm.wait()
    print(f"Датчик {i} IP-адрес {ip_addr} активирован, \"Wee-wee-wee-wee\"")

async def trigger():
    await asyncio.sleep(5)
    print("Датчики зафиксировали движение")
    alarm.set()

async def main():
    sensors = [sensor(i, ip[i]) for i in range(5)]
    await asyncio.gather(trigger(), *sensors)

asyncio.run(main())
```
