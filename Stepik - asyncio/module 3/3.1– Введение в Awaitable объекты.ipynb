## Пример 1. Корутина и Task

```python
import asyncio

# Определение корутины
async def my_coroutine():
    print("Запуск корутины")
    await asyncio.sleep(1)
    print("Завершение корутины")
```

```python
async def main():
    task = asyncio.create_task(my_coroutine())
    await task

asyncio.run(main())
```

## Пример 2. Несколько корутин и asyncio.gather

```python
import asyncio

async def cook_dish(n):
    print(f"Повар {n} начинает готовить")
    await asyncio.sleep(n)
    print(f"Повар {n} закончил готовить")
    return f"Блюдо от повара {n}"
```

```python
async def main():
    tasks = [asyncio.create_task(cook_dish(n)) for n in range(1, 4)]
    results = await asyncio.gather(*tasks)
    print(results)

asyncio.run(main())
```
