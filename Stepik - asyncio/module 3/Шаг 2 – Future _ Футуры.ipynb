## Callback-функции у Future

```python
import asyncio

def callback(future):
    print(f"Обратный вызов: результат = {future.result()}")

async def do_some_work(x):
    print(f"Выполнение некоторой работы: {x}")
    await asyncio.sleep(x)
    return x * 2

async def main():
    task = asyncio.ensure_future(do_some_work(2))
    task.add_done_callback(callback)
    await task

asyncio.run(main())
```

## ensure_future vs create_task

```python
import asyncio

async def my_coroutine(arg):
    await asyncio.sleep(1)
    print(f"Корутина my_coroutine создана с помощью {arg}")

async def main():
    future = asyncio.ensure_future(my_coroutine('ensure_future'))
    task = asyncio.create_task(my_coroutine('create_task'))

    print(f"ensure_future создаёт объект с типом {type(future)}")
    print(f"create_task создаёт объект с типом {type(task)}")

    await future
    await task

asyncio.run(main())
```
