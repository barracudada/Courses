## Базовый пример использования await

```python
import asyncio

async def main():
    print('Начинаем')
    await asyncio.sleep(1)
    print('Закончили')

asyncio.run(main())
```

## Последовательное ожидание корутин

```python
import asyncio

async def fetch_data(url):
    await asyncio.sleep(1)
    return f"Data from {url}"
```

```python
async def main():
    data1 = await fetch_data('https://www.example1.com')
    data2 = await fetch_data('https://www.example2.com')
    print(data1)
    print(data2)

asyncio.run(main())
```

## Параллельное выполнение с asyncio.gather

```python
import asyncio

async def cook_pasta():
    print('Начинаем готовить пасту')
    await asyncio.sleep(5)
    print('Паста готова')

async def cook_sauce():
    print('Начинаем готовить соус')
    await asyncio.sleep(3)
    print('Соус готов')
```

```python
async def main():
    await asyncio.gather(cook_pasta(), cook_sauce())

asyncio.run(main())
```
