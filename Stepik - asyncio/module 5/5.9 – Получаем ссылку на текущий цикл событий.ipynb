## asyncio.get_event_loop

```python
import asyncio

async def my_task(n):
    await asyncio.sleep(1)
    print(f'Task {n} done')

async def main():
    loop = asyncio.get_event_loop()
    tasks = [loop.create_task(my_task(i)) for i in range(3)]
    await asyncio.gather(*tasks)

asyncio.run(main())
```
