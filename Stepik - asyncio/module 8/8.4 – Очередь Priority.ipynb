## Очередь с приоритетом (asyncio.PriorityQueue)

```python
import asyncio

async def main():
    queue = asyncio.PriorityQueue()
    await queue.put((2, 'Средний приоритет'))
    await queue.put((1, 'Высокий приоритет'))
    await queue.put((3, 'Низкий приоритет'))

    while not queue.empty():
        priority, task = await queue.get()
        print(priority, task)

asyncio.run(main())
```
