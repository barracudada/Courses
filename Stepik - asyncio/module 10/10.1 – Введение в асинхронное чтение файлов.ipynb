## Синхронное чтение файла (блокирующее)

```python
def read_file_sync(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        return file.read()

# content = read_file_sync('example.txt')
# print(content)
```

## Асинхронное чтение файла

```python
import asyncio
import aiofiles

async def read_file_async(filename):
    async with aiofiles.open(filename, mode='r', encoding='utf-8') as file:
        content = await file.read()
        return content

# content = asyncio.run(read_file_async('example.txt'))
# print(content)
```
