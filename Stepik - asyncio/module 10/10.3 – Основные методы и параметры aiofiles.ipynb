## aiofiles.open — основные параметры

```python
import asyncio
import aiofiles

async def read_file(path):
    async with aiofiles.open(
        path,
        mode='r',
        encoding='utf-8',
        buffering=-1
    ) as f:
        return await f.read()

# asyncio.run(read_file('example.txt'))
```

## Асинхронная запись в файл

```python
import asyncio
import aiofiles

async def write_file(path, text):
    async with aiofiles.open(path, mode='w', encoding='utf-8') as f:
        await f.write(text)

# asyncio.run(write_file('output.txt', 'Hello aiofiles'))
```
