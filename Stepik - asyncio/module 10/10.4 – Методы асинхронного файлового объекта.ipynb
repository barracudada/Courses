## Основные методы асинхронного файлового объекта

```python
import asyncio
import aiofiles

async def file_methods_demo(path):
    async with aiofiles.open(path, 'w+', encoding='utf-8') as f:
        await f.write('Line 1\nLine 2\nLine 3')
        await f.seek(0)
        print(await f.readline())
        print(await f.readlines())
        print('Position:', await f.tell())

# asyncio.run(file_methods_demo('demo.txt'))
```

## flush, truncate, close

```python
import asyncio
import aiofiles

async def advanced_methods(path):
    async with aiofiles.open(path, 'w+', encoding='utf-8') as f:
        await f.write('Hello world')
        await f.flush()
        await f.truncate(5)

# asyncio.run(advanced_methods('demo2.txt'))
```
