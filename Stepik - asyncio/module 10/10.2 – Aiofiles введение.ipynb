## Чтение файла построчно с aiofiles

```python
import asyncio
import aiofiles

async def read_lines(filename):
    async with aiofiles.open(filename, 'r', encoding='utf-8') as f:
        async for line in f:
            print(line.strip())

# asyncio.run(read_lines('example.txt'))
```

## Асинхронная запись в файл

```python
import asyncio
import aiofiles

async def write_file(filename, text):
    async with aiofiles.open(filename, 'w', encoding='utf-8') as f:
        await f.write(text)

# asyncio.run(write_file('output.txt', 'Hello, async world!'))
```
