# Примеры IO bound и CPU bound

## IO bound пример

```
import time
import urllib.request

urls = [
    'http://www.python.org',
    'http://www.python.org/about/',
]

start_time = time.time()

for url in urls:
    print(f'Загрузка {url}')
    response = urllib.request.urlopen(url)
    print(f'Загружено {url}')

end_time = time.time()
print(f'Загружено {len(urls)} URL за {end_time - start_time} секунд')
```

## CPU bound пример

```
import time

def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

start_time = time.time()

for i in range(1, 11):
    print(f'Факториал {i} = {factorial(i)}')

end_time = time.time()
print(f'Вычислено за {end_time - start_time} секунд')
```
