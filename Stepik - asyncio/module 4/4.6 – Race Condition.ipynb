## Race Condition без синхронизации

```python
import asyncio

bank_account = 1000

async def withdraw(amount):
    global bank_account
    if bank_account >= amount:
        await asyncio.sleep(1)
        bank_account -= amount
```

```python
async def main():
    await asyncio.gather(withdraw(900), withdraw(900))
    print(bank_account)

asyncio.run(main())
```

## Решение с asyncio.Lock

```python
import asyncio

bank_account = 1000
lock = asyncio.Lock()

async def withdraw(amount):
    global bank_account
    async with lock:
        if bank_account >= amount:
            await asyncio.sleep(1)
            bank_account -= amount
```

```python
async def main():
    await asyncio.gather(withdraw(900), withdraw(900))
    print(bank_account)

asyncio.run(main())
```
