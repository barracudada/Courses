## Проблема неатомарности

```python
import asyncio

shared_resource = 0

async def update_resource():
    global shared_resource
    temp = shared_resource
    await asyncio.sleep(1)
    shared_resource = temp + 1
```

```python
async def main():
    await asyncio.gather(update_resource(), update_resource())
    print(shared_resource)

asyncio.run(main())
```

## Решение с asyncio.Lock

```python
import asyncio

shared_resource = 0
lock = asyncio.Lock()

async def update_resource():
    global shared_resource
    async with lock:
        temp = shared_resource
        await asyncio.sleep(1)
        shared_resource = temp + 1
```

```python
async def main():
    await asyncio.gather(update_resource(), update_resource())
    print(shared_resource)

asyncio.run(main())
```
