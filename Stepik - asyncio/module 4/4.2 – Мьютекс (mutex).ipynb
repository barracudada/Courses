## Использование asyncio.Lock (мьютекс)

```python
import asyncio

async def access_resource(resource, lock):
    async with lock:
        print(f"Доступ получен {resource}")
        await asyncio.sleep(1)
        print(f"Доступ завершён {resource}")
```

```python
async def main():
    lock = asyncio.Lock()
    resource = '--Защищаемый ресурс--'
    tasks = [access_resource(resource, lock) for _ in range(5)]
    await asyncio.gather(*tasks)

asyncio.run(main())
```
