## Асинхронный контекстный менеджер

```python
import asyncio

class AsyncResource:
    async def __aenter__(self):
        print('Ресурс получен')
        await asyncio.sleep(0.5)
        return self

    async def __aexit__(self, exc_type, exc, tb):
        print('Ресурс освобождён')
        await asyncio.sleep(0.5)

async def main():
    async with AsyncResource():
        print('Работаем с ресурсом')
        await asyncio.sleep(1)

asyncio.run(main())
```

## async with и asyncio.Lock

```python
import asyncio

async def worker(lock, i):
    async with lock:
        print(f'Task {i} вошла в критическую секцию')
        await asyncio.sleep(1)

async def main():
    lock = asyncio.Lock()
    await asyncio.gather(*(worker(lock, i) for i in range(3)))

asyncio.run(main())
```
