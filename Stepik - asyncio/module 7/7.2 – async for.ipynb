## Асинхронный генератор

```python
import asyncio

async def agen():
    for i in range(5):
        await asyncio.sleep(0.5)
        yield i

async def main():
    async for value in agen():
        print(value)

asyncio.run(main())
```

## async for + список-выражение

```python
import asyncio

async def agen():
    for i in range(3):
        await asyncio.sleep(0.2)
        yield i

async def main():
    result = [i * i async for i in agen()]
    print(result)

asyncio.run(main())
```
