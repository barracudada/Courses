```python
# Загружаем набор данных BostonHousing
!wget https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv
```

```python
# Импортируем Векторное представление фичей VectorAssembler из pyspark и Линейную Регрессию (LinearRegression)
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression

# Читаем из директории наор данных BostonHousing.csv в SparkDataFrame
dataset = spark.read.csv('BostonHousing.csv', inferSchema=True, header =True) 

# Выводим первые 20 строк из набора данных
dataset.show()
```

```python
# Применяем фильтр к набору данных
dataset.where('zn > 0').show() # что эквивалентно SQL select * from dataset where zn >0
```

```python
#SQL
dataset.registerTempTable("train_data") # в качестве шаблона, на котором будем смотреть запросы SQL загружаем обучающую выборку из dataset
spark.sql("select * from train_data where zn>0 and age like '%4.%'").show() # в качестве аргумента используем SQL запрос
```

```python
# Сгруппируем по заданным характеристикам
spark.sql("select rad, count(age),mean(age) from train_data group by rad ").show()
```

```python
# выводим схему набора данных
dataset.printSchema()
```

```python
# Объединяем все числовые переменные в одну векторную колонку
assembler = VectorAssembler(
    inputCols=['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat'],
    outputCol = 'Attributes') # создаем объект, преобразующий ряды и столбы в характеристические вектора
output = assembler.transform(dataset) # преобразуем 
output
```

```python
output.show()
```

```python
# Выбираем X (созданная нами векторная колонка Attributes) и Y(колонка medv)
finalized_data = output.select("Attributes","medv")
finalized_data
```

```python
# Выводим подлготовленные для обучения модели данные
finalized_data.show(truncate=False)
```

```python
#Делим данные на train/test 80:20
train_data,test_data = finalized_data.randomSplit([0.8,0.2])

# Инициализируем линейную регрессию указываем колонку с фичами и таргет переменной
regressor = LinearRegression(featuresCol = 'Attributes', labelCol = 'medv')

#Запускаем обучение модели
regressor = regressor.fit(train_data)

#Проверяем точность модели на тестовых данных
pred = regressor.evaluate(test_data)
pred.predictions
```

```python
#Выводим результат
pred.predictions.show()
```

```python
# Импортируем RegressionEvaluator для проверки качества модели
from pyspark.ml.evaluation import RegressionEvaluator

eval_r = RegressionEvaluator(labelCol="medv", predictionCol="prediction", metricName="mae")

# Считаем и выводим Root Mean Square Error
mae = eval_r.evaluate(pred.predictions)
print("MAE: %.3f" % mae)
```

```python
# Сохраняем Spark DataFrame в формате Parquet
pred.predictions.write.mode("overwrite").format("parquet").save("predictions.csv")
```

```python
# Конвертируем Spark DataFrame в Pandas
df_pred = pred.predictions.toPandas()
df_pred.head()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('DL with Spark Deep Coqnition').getOrCreate()
sc = spark.sparkContext
```

```python
from pyspark.ml.image import ImageSchema
image_df = ImageSchema.readImages('flower_photos/sample/')
image_df.show(truncate=True)
```

```python
# Это будет наша Keras модель
from keras.applications import InceptionV3

model = InceptionV3(weights="imagenet")
model.save('model-full.h5')  
```

```python
#### Аналогичный разворот модели предобученной сетки
from keras.applications.inception_v3 import preprocess_input
from keras.preprocessing.image import img_to_array, load_img
import numpy as np
from sparkdl import KerasImageFileTransformer

def loadAndPreprocessKerasInceptionV3(uri_df):
  # загрузка изображения с помощью инструментов keras
    image = img_to_array(load_img(uri_df, target_size=(299, 299)))  # размеры изображения для InceptionV3
    image = np.expand_dims(image, axis=0)
    return preprocess_input(image)

transformer = KerasImageFileTransformer(inputCol="uri", outputCol="predictions", # uri - идентификаторы будущего xTrain
                                        modelFile='model-full.h5',  # здесь путь к модели
                                        imageLoader=loadAndPreprocessKerasInceptionV3, # объект-предобработка/загрузка изображения
                                        outputMode="vector") # по умолчанию

fs = !ls flower_photos/sample/*.jpg
```

```python
from pyspark.sql.types import StringType
uri_df = spark.createDataFrame(['flower_photos/sample/10140303196_b88d3d6cec.jpg'], StringType()).toDF("uri") # возьмем один экземпляр и склеим из него DataFrame
uri_df.show()
```

```python
keras_pred_df = transformer.transform(uri_df)
```
