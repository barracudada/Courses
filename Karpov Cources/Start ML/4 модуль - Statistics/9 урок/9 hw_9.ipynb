### 1

Что происходит с ложноположительной ошибкой, если оценить прокрас метрики в эксперименте каждый день по всем предыдущим дням?

- [+] Увеличивается выше выбранного уровня значимости
- Стремится к нулю
- Остаётся на выбранном уровне значимости

**Комментарий**

Подсчитывать метрики каждый день похоже на то, чтобы проводить больше экспериментов, то есть мы сталкиваемся с проблемой множественной проверки гипотез.

Можно помнить, что "если проверять бесконечно число гипотез, то некоторые, да прокрасятся". В этом случае тоже ложноположительная ошибка вырастает.

### 2

Что происходит при оценке значимости каждый день на основе всех прошедших дней эксперимента?

- Нарушается независимость наблюдений
- Нарушаются условия применимости статистического критерия
- [+] Ситуация, схожая с множественной проверкой гипотез

**Комментарий**

Мы как будто проверяем много гипотез "прокрасилось ли за первый день", "прокрасилось ли за первые два дня", ...

### 3

Какие из периодов можно считать репрезентативными для проведения эксперимента (особенно на фоне остальных вариантов ответа)?

- С 28 апреля по 13 мая
- С 1 июня по 4 июня
- [+] С 1 августа по 31 августа
- [+] С 7 октября по 13 октября

**Комментарий**

Период с апреля по май затрагивает майские праздники, в которые пользователи ведут себя иначе — едут в отпуска и на дачи, например. С 1 по 4 июня просто нерепрезентативный период, так как не включает в себя полную неделю.

А два оставшихся периода вполне нормальные.

### 4

Представим, что мы тестируем новый алгоритм переводов в приложении-переводчике. Мы решили забивать пользователей на 2 группы по 50%. Для этого при каждом входе пользователя в приложении мы «подкидываем монетку» и решаем, в какую группу отправить пользователя. Эксперимент будет длиться 2 недели.

Нормальный ли это подход?

**Решение**

Это опасно, так как один и тот же пользователь за период эксперимента может зайти в приложение несколько раз и оказаться в разных группах. Это нарушит корректность A/B тестирования, так как сломается независимость наблюдений между группами.

### 5

Представим, что мы тестируем новый алгоритм переводов в приложении-переводчике. Мы решили забивать пользователей на две группы по 50%. Для этого при каждом входе пользователя в приложении мы вычисляем остаток от деления на 2 после применения хэш-функции с солью к идентификатору пользователя, чтобы определить, в какую группу отправить пользователя. Эксперимент будет длиться 2 недели.

Нормальный ли это подход?

**Решение**

А такой подход как раз корретный, если нет проблем с хэш-функцией.

### 6

Из каких соображений нужно определять длительность эксперимента?

- столько, чтобы точно прокрасить нашу метрику
- [+] покрыть полный цикл активности (например, неделю)
- [+] собрать достаточно наблюдений для ожидаемого эффекта
- как можно короче — если можно за день, то лучше за день

**Комментарий**

Неправильно собирать столько, чтобы прокрасилась наша метрика — вдруг изменения никакого нет. Да и зачем нам проводить эксперимент, если цель именно прокрасить метрику? :)

Важно собирать репрезентативные данные (полный цикл пользователей), достаточно наблюдений с точки зрения ожидаемого эффекта. 

Совсем мало данных вряд ли хорошая идея, в том числе из-за репрезентативности данных.

### 7

В чём основная проблема измерения большого количества метрик в эксперименте?

- [+] Среди большого количества метрик легко найти положительное улучшение в эксперименте и получить ложноположительную ошибку
- Подсчитывать большое количество метрик затратно
- У большого количества метрик нет проблемы — можем делать более сложные выводы

**Комментарий**

Подсчитывать много метрик хоть и сложнее, но на деле обычно это просто цикл по метрикам.

Среди большого количества метрик сильно проще что-то прокрасить, ошибка первого рода вырастает, надо быть аккуратными со своими выводами.

### 8

Какие требования предъявляются к функции для разбиения пользователей на группы?

- [+] репрезентативность групп пользователей
- [+] воспроизводимость (повторное применение даёт тот же результат)
- [+] равномерность групп пользователей

**Комментарий**

Воспроизводимость — каждый раз пользователь должен попадать в ту же группу в рамках эксперимента, чтобы не нарушать независимость

Равномерность групп пользователей — группы должны получаться нужных размеров и равными между собой

Репрезентативность групп пользователей — разбиение должно делать такие группы, чтобы состав пользователей в них по разным факторам был примерно одинаковым
